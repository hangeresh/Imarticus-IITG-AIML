## Numeric Statistic - Correlation then Plot as Heatmap

```python
# Dataframe correlation
cor = tips.corr()
print(cor)
# Plot heatmap of the correlation matrix with Seaborn
sns.heatmap(cor, cmap="Blues", annot=True)
```
